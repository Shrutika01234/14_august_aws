```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

### ***MICRO TASK 1***

```python
df=pd.read_csv('EOD-AAPL.csv')
df
```

```python
df.describe()
```

```python
df.dtypes
```

```python
df.columns
```

```python
df['PCT_change']=((df['Adj_Close']-df['Adj_Open'])/df['Adj_Open']*100)
df
```

```python
df['HL_PCT']=((df['Adj_High']-df['Adj_Low'])/df['Adj_Close']*100)
df
```

```python
df.fillna(-99999)
```

### ***MICRO TASK 2***

```python
import math
new=pd.DataFrame()
new= df[['Adj_Close','PCT_change','HL_PCT','Adj_Volume']]
new.fillna(value=-99999, inplace=True)
new
```

```python
var1= int(math.ceil(0.01 * len(df)))
new['label'] = df['Adj_Close'].shift(var1)
new.dropna(inplace=True)
new
```

### ***MICRO TASK 3***

```python
#importing the needed libraries
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler
```

```python
x = new[['Adj_Close','PCT_change','HL_PCT','Adj_Volume']]
x = preprocessing.scale(x)
y = new['label']
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
reg = LinearRegression()  
reg.fit(X_train, y_train) #training the algorithm
```

```python
print("INTERCEPT IS :",end=" ")
print(reg.intercept_)
print("COEFFICIENTS ARE :",end=" ")
print(reg.coef_)
```

```python
score1 = reg.score(X_test, y_test)
score1
```

```python
y_pred = reg.predict(x)
y_pred
```

```python
print("THE FIRST TEN PREDICTIONS ARE : ",end=" ")
df_1 = pd.DataFrame()
df_1['ACTUAL'] = new['label']
df_1['PREDICTED'] = y_pred
df_1.head(10)
```

```python
ax1 = sns.distplot(y, hist=False, color="r", label="Actual Value of label")
sns.distplot(y_pred, hist=False, color="b", label="Fitted Values of Pred" , ax=ax1)
```
